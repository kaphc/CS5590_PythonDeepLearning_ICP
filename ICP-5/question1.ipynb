Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import dataset

```python
data = pd.read_csv('data/train.csv')
data.head(10)
```

define graph

```python
plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (20, 15)
```

plot the data

```python
garage_field = data['GarageArea']
sale_price = data['SalePrice']
plt.scatter(garage_field, sale_price, alpha=.70, color='r')
plt.xlabel('Garage_Field:Column')
plt.ylabel('Sale_Price:Target')
plt.show()
```

describe GarageArea

```python
data['GarageArea'].describe()
```

describe SalePrice

```python
data['SalePrice'].describe()
```

filter outliers

```python
data_without_outliers = data['GarageArea'] <= 800
data = data[data_without_outliers]
data_without_outliers = data['GarageArea'] > 260
data = data[data_without_outliers]
data_without_outliers = data['GarageArea'] != 0
data = data[data_without_outliers]
```

plot data without outliers

```python
garage_field = data['GarageArea']
sale_price = data['SalePrice']
plt.scatter(garage_field, sale_price, alpha=.70, color='g')
plt.xlabel('Garage_Field:Column')
plt.ylabel('Sale_Price:Target')
plt.show()
```
